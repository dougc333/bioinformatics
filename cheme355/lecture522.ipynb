of well defined protein which we have crystal structure. Goal here is to get increased tolerance
to solvent, DMF. Not something which you bind w/enzymes. What do you want to change in enzyme? Surface polarity you can 
change, may allow protein to fold better. Protease in context of organic solvent. If increase speed of reaction
will this make tolerance better? Yes, increase Kcat. 
1) diversity used error prone PCR, get 3bp mutations per gene. You only want 1 AA mutation. so 1 codon or 3bp mutation.

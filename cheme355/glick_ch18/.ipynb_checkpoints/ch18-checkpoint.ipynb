Ti(Ti = tumor inducing plasmid) plasmid systems for plants, these are T-DNA systems where there are 2 plasmids, a binary shuttle vector which can 
live in ecoli or agrobacterioum(the gram negative bacterium that causes crown gall tumors, a big 
wart on the outside of a plant) and a helper plasmid. 
Gram negative soil bacterium Agrobacterium tumefaciens only infects certain plants, dicots. some grapes. The bacterium
transfers the T-DNA from the bacterium into the plant genome. 
When a plant is injured it releases chemoattractants which attract agrobacterioum. Acetyosurgione is released by the 
injured plant which binds to the agrobacterium on a site which is encoded by virA receptor. The virA receptors 
are proteins which exist as a dimer.

T-DNA contains the 4 genes, iaaM, iaaH, tmr which are responsible for crown gall formation and a gene for opine.
Gene iaaM  codes for the enzyme tryptophan 2-monooxygenase, which converts tryptophan to indole 3-acetamide, 
Gene iaaH  encodes indole 3-acetamide hydrolase, which converts indole 3-acetamide to indoleacetic acid (Fig. 18.5A). 
In addition, the T-DNA region carries the tmr gene (also known as ipt ), 
which encodes isopentenyltransferase. This enzyme adds an isopentenyl side chain to 
5 ′-AMP to form isopentenyladenosine 5 ′-phosphate, the first committed step in the synthesis of the 
cytokinin isopentenyladenine (Fig. 18.5B). 
Cytokinins transzeatin and transribosylzeatin. Auxin and these cytokinins regulate cell growthn and cause plant
crown galls. 
<li>opines are unique and condensation products of either an amino acid and keto acid or an AA and sugar. </li>

inside the crown gall and secreted. They can by used as C and N source by A tumefaciens cell that carries a Ti plasmid
gene for catabolism of opine. teh catabolism gene is on Ti plasmid and NOT part of T-DNA region!!!! A tumerfiens has its

<h6>Ti Plasimd Systems</h6>
Problems with Ti plasmid/A tumefaciens to deliver genes into plants.
<li>Production of phytohormones by transformed cells growing in culture prevents them form being
regenerated into plants. Auxin and cytokinin genes have to be removed from Ti plasmid vector. </li>
<li>Opine synthesis genes not useful in transgenic plant and lowers yield. Remove. </li>
<li>Ti plasmids are large at 200-800kb. make smaller</li>
<li>Ti plasimid does not replicate inside EColi, the modified one below does replicate in EColi</li>
<li>Transfer of TDNA does not always end at left border of TDNA plasmid. </li>
Use recombinant DNA tech to create more Ti Plasmid vectors:
    <li>SElectable marker, nemoycin phosphotransferase, KAN gene. put under control of plant gene promoter</li>
    <li>ORI, allows plasmid to replicate in EColi</li>
    <li>right border sequence in TDNA</li>
    <li>polylinker(multiple cloning site), allows user to add gene in TDNA section of plasmid</li>
    <li>killer gene encoding toxin downstream from left border to prevent unwanted vector DNA past left border from
    beind incorporated into transgenic plants</li>
    <li>These cloning vectors lack vir genes so these plasmids cant be incorported into plants.</li>

<li>2.  How do (1) binary and (2) cointegrate Ti plasmid-based vector systems for plant transformation differ from one another? </li>
1)binary vectors, there are 2 separate plasmids one containing T-DNA and another containing vir genes. These
are not on the same plasmid. Provided one of each enters and recombines in the host cell this works. These 2 separate plasmids replicate in bacteria first. 
2) cointegrate Ti Plasmids are 2 plasimids, one with gene insert into disarmed Ti plasmid(cancer genes out) and
the Ecoli TDNA plamsid, use homologous recombination inside agrobacterium to create single plasmid. For HR to occur the Ecoli plasmid and disarmed Ti plasmid have to have common sequence.

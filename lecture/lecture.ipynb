```python
# 
import os
print(os.getcwd())
words=[]
with open("../TRAVELSTORIES","r") as file_handle:
    for line in file_handle:
        print(line)
        words.append(line.split())
```

```python
print(words)
flattened = [x for sublist in words for x in sublist]
```

```python
print(flattened)
```

```python
#create dictionary, lower case and turn the list into a set
#sorted set/tree faster for search
lowercase = [x.lower() for x in flattened]
```

```python
print(lowercase)
```

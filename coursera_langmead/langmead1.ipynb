```python
def longest_common_prefix(s1,s2):
    idx = 0
    #would be better to use s1[idx]==s2[idx]
    while(idx < len(s1) and idx < len(s2) and s1[idx]==s2[idx]):
        idx+=1
    return s1[:idx]
#careful w/indexes if you do s1[:idx] == s2[:idx] then you are off by 1 index need to return s1[:idx-1]
print(longest_common_prefix("ACGCT","ACGCTA"))
```

```python
complement = {'A':'T', 'C':'G','G':'C','T':'A'}
def reverse_complement(s):
    c=[]
    for x in s:
        c.append(complement[x])
    return ''.join(c)[::-1]
print(reverse_complement("ATCG"))

#the better design is the prepend the complement and to use a string instead of a list
def reverse_c2(s):
    t=''
    for base in s:
        t = complement[base]+t
    return t
print(reverse_c2("ATCG"))
```

```python
#example of exact matching
def read_genome(filename):
    genome = ''
    with open(filename,'r') as fh:
        for line in fh:
            genome = genome+line.rstrip() #make sure to append
    return genome

genome = read_genome('lambdaphage.txt')
print(len(genome))

def match(s,genome):
    '''
    input: s to match to genome
    output: return list of positions of where s is in genome
    '''
    pos = []
    for idx_pos in range(0,(len(genome)-len(s)+1)):
        if s==genome[idx_pos:idx_pos + len(s)]:
            pos.append(idx_pos)
    return pos

print (match("AAA","ACTAAAG"))
```
